```python
class Calculator:
	def __init__(self):
		self.result = 0
	
	def add(self, num):
		self.result += num
		return self.result
	
	def sub(self, num):
		self.result -= num
		return self.result
	
	def mul(self, num):
		self.result *= num
		return self.result
	
	def div(self, num):
		self.result /= num
		return self.result
```

```python
cal3 = Calculator()
print(cal3.add(10))
print(cal3.sub(5))
```

```python
# 객체에 연산할 숫자 지정하기

class FourCal:
	def setdata(self, first, second):
		self.first = first
		self.second = second
	
	def add(self):
		result = self.first + self.second
		return result
	
	def sub(self):
		result = self.first - self.second
		return result
	
	def mul(self):
		result = self.first * self.second
		return result
	
	def div(self):
		result = self.first / self.second
		return result


cal3 = FourCal()
print(type(cal3))
cal3.setdata(4, 2)

print(cal3.first)
print(cal3.second)
print(cal3.add())
print(cal3.sub())
print(cal3.mul())
print(cal3.div())
```

```python
class FourCal:
	def __init__(self, first, second):
		self.first = first
		self.second = second

	def add(self):
		result = self.first + self.second
		return result
	
	def sub(self):
		result = self.first - self.second
		return result
	
	def mul(self):
		result = self.first * self.second
		return result
	
	def div(self):
		result = self.first / self.second
		return result

cal4 = FourCal(10, 20)
print(cal4.first)
print(cal4.second)

print(cal4.add())
print(cal4.sub())
print(cal4.mul())
print(cal4.div())
```

```python
class MoreFourCal(FourCal):
	def pow(self):
		result = self.first ** self.second
		return result
```

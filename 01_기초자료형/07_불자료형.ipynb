```python
# 불 자료형은 어떻게 사용할까?

a = True
b = False

print(a)
print(type(a))
print(b)
print(type(b))
```

```python
print(1 == 1)
print(2 > 1)
print(2 < 1)
```

```python
# 값이 없는 경우나 0인 경우만 False
print(1, bool(1))
print(0, bool(0))
print(2, bool(2))
print("python", bool("python")) 
print("", bool(""))
print("[]", bool([]))
print("[1, 2, 3]", bool([1, 2, 3]))
```

```python
if [1, 2, 3]:
	print("참")
else:
	print("거짓")
```

```python
# 논리연산자
# 1. and 연산자
# 하나라도 False가 있으면 False

print(True and True)
print(True and False)
print(False and False)
print(True and True and False)

# 2. or 연산자
# 하나라도 True가 있으면 True
print(True or True)
print(True or False)
print(False or False)
print(False or False or True)

# 3. not 연산자

print(not True)
print(not False)
print(not 1)
print(not [])
```

```python
# 문자열 포매팅
# 1. 숫자 바로 대입
print("I eat %d apples" % 3)

# 2. 문자열 바로 대입
print("I eat %s apples" % "five")

# 3. 숫자 값을 나타내는 변수로 대입
number = 3
print("I eat %d apples" % number)

# 4. 2개 이상의 값 넣기
number = 10
day = "three"

print("I ate %d apples, so i was sick for %s days" % (number, day))
```

### 문자열 포맷 코드
- %s : 문자열(string)
- %c : 문자 1개(character)
- %d : 정수(integer)
- %f : 실수(float)
- %o : 8진수
- %x : 16진수
- %% : 문자 % 그 자체 

```python
# format 함수를 사용한 포매팅
# 1. 숫자 바로 대입
print("i eat {0} apples".format(3))

# 2. 문자열 바로 대입하기
print("i eat {0} apples".format("five"))

# 3. 2개 이상의 값 넣기
number = 10
day = "three"

print("I ate {0} apples, so I was sick for {1}".format(number, day))
```

```python
# f문자열 포매팅
name = "홍길동"
age = 30

print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.")
```

```python
# 문자열 관련 함수들
# 내장함수: 자체적으로 설치되있는 함수
# print(), len(), type() ...

# 메서드: 객체가 따로 있어야 사용가능한 함수
# .count(), .find(), .index(), .join(), .upper(), .lower(), .strip()
# .lstrip(), .rstrip(), .replace(), .split()

# 1. 문자 개수 세기 - count
a = "hobby"
print(a.count("b"))

# 2. 위치 알려 주기 - find
print(a.find("b"))
print(a.find("k"))

# 3. 위치 알려 주기 - index
print(a.index("b"))

# 4. 문자열 삽입 - join
print(",".join("abcd"))

# 5. 소문자를 대문자로 바꾸는 함수 - upper
a = "hello"
print(a.upper())

# 6. 대문자를 소문자로 바꾸는 함수 - lower
a = "HI"
print(a.lower())

# 7. 공백지우기 - strip
a = "   hi   "
print(a)
print(a.strip())

# 8. 왼쪽 공백지우기 - lstrip
# 9. 오른쪽 공백지우기 - rstrip

print(a.lstrip())
print(a.rstrip())

# 10. 문자열 바꾸기 - replace
a = "Life is too short"
print(a.replace("Life", "Your leg"))

# 11. 문자열 나누기 - split
a = "Life is too short"
print(a.split())
```

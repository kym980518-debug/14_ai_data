```python
# 집합자료형은 어떻게 만들까?

a = set([1, 2, 3])
print(type(a))
print(a)
```

```python
# set은 중복 x, 순서 x

b = set("hello")
print(b)
```

```python
c = {1, 2, 3}
print(type(c))
print(c)
```

```python
d = {"a", "b", "c"}
print(d)
```

```python
# 교집합 - intersection / &

s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

print(s1.intersection(s2))
print(s1 & s2)
```

```python
# 합집합 - union / |

s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

print(s1.union(s2))
print(s1 | s2)
```

```python
# 차집합 - difference(앞 인수에서 뒤 인수를 뺌) / -

s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

print(s1.difference(s2))
print(s2.difference(s1))
print(s1 - s2)
print(s2 - s1)
```

```python
# 값 1개 추가하기 - add

s1 = set([1, 2, 3])
s1.add(4)
print(s1)
```

```python
# 값 여러 개 추가하기 - update

s1 = set([1, 2, 3])
s1.update([4, 5, 6])
print(s1)
```

```python
# 특정 값 제거하기 - remove

s1 = set([1, 2, 3])
s1.remove(2) #( ) 안에 인덱스가 아니라 특정 값 적용
print(s1)
```

```python
# 모든 값 제거하기 - clear

s1 = set([1, 2, 3])
s1.clear()
print(s1)
```
